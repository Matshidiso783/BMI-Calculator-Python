```python
#BMI Calculator

##Project Overwview
This is a python project that calculates Body Mass Index (BMI) baesd on user input for weight and height.

##How it works
-User enters wewight(kg)
-User enters height(m)
-The program calculates BMI using the formula:
-The result is then classified intio:
    -underweight
    -Normal weight
    -Overweight
    -Obesity
    -Severely Obese
    -Morbidly obese

## Skills Demonstrated
- python input/outout
-Conditional Statements (if/elif/else)
-Basic mathematical operations

##Purpose
This project was built to practicePython basiccs and logical thninking.
```

```python
name = input("Enter your name: ")

weight = int(input("enter your wieght in pounds: "))

height= int(input(" enter your height in inches: "))

BMI = (weight* 703) / (height * height)
print(BMI)


if BMI<=0:
    print(name +", enter valid inputs")
elif BMI<18.5:
      print (name + ", you are underweight. ")
elif BMI <=24.9:
     print(name + ", you are normal weight.")
elif BMI<=29.9:
    print(name + ", you are overweight.")
elif BMI<=34.9:
     print(name + ", you are obese.")
elif BMI<=39.9:
    print(name + ", severely obese.")
else:
     print(name +", you are morbidly obese")
```
